```python
import numpy as np

words = [
    "low",
    "lower",
    "lowest",
    "new",
    "newer",
    "widest"
]

words = [word.lower().strip() for word in words]

print("========== Corpus ==========")
print(words)

vocabulary = {}

for word in words:

    token = " ".join(list(word)) + " </w>"

    vocabulary[token] = vocabulary.get(token, 0) + 1

print("\n========== Initial Vocabulary ==========")

for token, count in vocabulary.items():
    print(token, ":", count)

frequency = np.array(list(vocabulary.values()))

print("\nWord Frequencies :", frequency)
print("Total Words      :", np.sum(frequency))
print("Maximum Frequency:", np.max(frequency))


def get_pair_counts(vocabulary):

    pairs = {}

    for token, count in vocabulary.items():

        symbols = token.split()

        for i in range(len(symbols)-1):

            pair = (symbols[i], symbols[i+1])

            pairs[pair] = pairs.get(pair, 0) + count

    return pairs

def merge_pair(pair, vocabulary):

    updated_vocab = {}

    old = " ".join(pair)

    new = "".join(pair)

    for token, count in vocabulary.items():

        merged = token.replace(old, new)

        updated_vocab[merged] = count

    return updated_vocab


merge_rules = []

num_merges = 10

for step in range(num_merges):

    pair_counts = get_pair_counts(vocabulary)

    if len(pair_counts) == 0:
        break

    best_pair = max(pair_counts, key=pair_counts.get)

    merge_rules.append(best_pair)

    print("\n------------------------")
    print("Merge", step + 1)
    print("Best Pair :", best_pair)
    print("Frequency :", pair_counts[best_pair])

    vocabulary = merge_pair(best_pair, vocabulary)

final_vocab = set()

for token in vocabulary:
    final_vocab.update(token.split())

print("\n========== Final Vocabulary ==========")
print(sorted(final_vocab))

print("\nVocabulary Size :", len(final_vocab))

def encode(word):

    tokens = list(word.lower()) + ["</w>"]

    for pair in merge_rules:

        i = 0

        while i < len(tokens)-1:

            if tokens[i] == pair[0] and tokens[i+1] == pair[1]:

                tokens = tokens[:i] + ["".join(pair)] + tokens[i+2:]

            else:
                i += 1

    return tokens

def decode(tokens):

    return "".join(tokens).replace("</w>", "")


test_words = [
    "newest",
    "lowest",
    "lower"
]

print("\n========== Testing ==========")

for word in test_words:

    encoded = encode(word)

    decoded = decode(encoded)

    print("\nOriginal :", word)
    print("Encoded  :", encoded)
    print("Decoded  :", decoded)

print("\n========== Learned Merge Rules ==========")

for i, rule in enumerate(merge_rules, start=1):
    print(i, ":", rule)

print("\n========== Program Completed ==========")
```
